#ML Project: Solubility Prediction using Molecular Properties

##Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##Load data

```python
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df.head()
```

## Data preparation

###Data separation as X and y

```python
y = df["logS"]
y
```

```python
X = df.drop('logS', axis=1)
X
```

##Data splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

##Model building

##Linear regression

###Training the model

```python
from sklearn.linear_model import LinearRegression

m = LinearRegression()
m.fit(X_train, y_train)
```

###Applying the model to make prediction

```python
y_train_pred = m.predict(X_train)
y_test_pred = m.predict(X_test)
```

```python
y_train_pred
```

```python
y_test_pred
```

###Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)
```

```python
res = pd.DataFrame(['Linear Regression', mse_train, r2_train, mse_test, r2_test]).transpose()
res.columns = ["Method", "mse_train", "r2_train", "mse_test", "r2_test"]
res
```

##Random Forest

###Training the model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

###Applying the model to make a prediction

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

###Evaluating the model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_mse_train = mean_squared_error(y_train, y_rf_train_pred)
rf_r2_train = r2_score(y_train, y_rf_train_pred)

rf_mse_test = mean_squared_error(y_test, y_rf_test_pred)
rf_r2_test = r2_score(y_test, y_rf_test_pred)
```

```python
res_rf = pd.DataFrame(['Random Forest', rf_mse_train, rf_r2_train, rf_mse_test, rf_r2_test]).transpose()
res_rf.columns = ["Method", "mse_train", "r2_train", "mse_test", "r2_test"]
res_rf
```

##Model comparision

```python
df_models = pd.concat([res, res_rf])
df_models.reset_index(drop=True)
```

##Data visualization

```python
z = np.polyfit(y_train, y_train_pred, 1)
p = np.poly1d(z)
```

```python
plt.figure(figsize=(5,5))
plt.xlabel("Experimental logS")
plt.ylabel("Predicted logS")

plt.scatter(x=y_train, y=y_train_pred, alpha=0.2)
plt.plot(y_train, p(y_train), color="red")
plt.plot()
```
